```python
import pandas as pd

df = pd.read_csv('TeluguMovies_dataset.csv')
print(df.head())
print(df.columns)
print(df.info())
```

```python
df['Genre_list'] = df['Genre'].str.split(',')
```

```python
df_exploded = df.explode('Genre_list')
```

```python
df_exploded['Genre_list'] = df_exploded['Genre_list'].str.strip()
```

```python
import matplotlib.pyplot as plt

df_exploded['Genre_list'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Genre Distribution')
plt.ylabel('Count')
plt.show()
```

```python
#Number of Movies Released Per Year
import matplotlib.pyplot as plt

df['Year'] = df['Year'].fillna(0).astype(int)  # Convert year to int (fill NaNs with 0)

year_counts = df[df['Year'] > 1900]['Year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
year_counts.plot(kind='line', marker='o')
plt.title('Number of Telugu Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
avg_rating_per_year = df.groupby('Year')['Rating'].mean().dropna()

plt.figure(figsize=(12,6))
avg_rating_per_year.plot(kind='line', color='orange', marker='o')
plt.title('Average Movie Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Pastel color palette for bars (10 colors for 10 movies)
pastel_colors = [
    "#AEC6CF",  # pastel blue
    "#FFB347",  # pastel orange
    "#77DD77",  # pastel green
    "#F49AC2",  # pastel pink
    "#FFD1DC",  # light pastel pink
    "#CBAACB",  # pastel purple
    "#FFFACD",  # pastel yellow
    "#B0E0E6",  # pastel teal
    "#E6E6FA",  # lavender
    "#FF6961"   # pastel red
]

top_movies = df.sort_values(by='No.of.Ratings', ascending=False).head(10)

plt.figure(figsize=(12,6))
bars = plt.barh(top_movies['Movie'], top_movies['No.of.Ratings'], color=pastel_colors[:10])
plt.gca().invert_yaxis()
plt.title('Top 10 Telugu Movies by Number of Ratings')
plt.xlabel('Number of Ratings')
plt.show()
```

```python
cert_counts = df['Certificate'].value_counts()

plt.figure(figsize=(8,5))
cert_counts.plot(kind='bar', color='coral')
plt.title('Certificate Distribution of Telugu Movies')
plt.xlabel('Certificate')
plt.ylabel('Number of Movies')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get top 4 certificates by count
top4_cert_counts = df['Certificate'].value_counts().head(4)

pastel_colors = [
    "#AEC6CF", "#FFB347", "#77DD77", "#F49AC2"
]

plt.figure(figsize=(6,6))
wedges, texts, autotexts = plt.pie(
    top4_cert_counts,
    labels=None,
    colors=pastel_colors,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.85,
    wedgeprops={'edgecolor': 'black'},
    textprops={'fontsize': 10}
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
plt.gca().add_artist(centre_circle)

plt.legend(
    wedges,
    top4_cert_counts.index,
    title="Certificates",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize=10
)

plt.title('Top 4 Movie Certificates Distribution', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_genre_wordcloud():
    all_genres = ', '.join(df['Genre'].dropna()).replace(',', '')
    
    if not all_genres.strip():
        print("No genre data found.")
        return
    
    wordcloud = WordCloud(
        width=800,
        height=400,
        background_color='white',
        colormap='Pastel1'
    ).generate(all_genres)
    
    plt.figure(figsize=(12, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Movie Genres', fontsize=16)
    plt.show()

# Call the function
generate_genre_wordcloud()
```
